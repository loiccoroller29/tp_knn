## Remplacement des valeurs qualitatives

```python
# On remplace les a/A par des 1
df = df.replace("a", 1)
df = df.replace("A", 1)

# On remplace les b/B par des 2
df = df.replace("b", 2)
df = df.replace("B", 2)

# On remplace les c/C par des 3
df = df.replace("c", 3)
df = df.replace("C", 3)
```

```python
df = df.replace(['a', 'b', 'c'], [1, 2, 3])
df = df.replace(['A', 'B', 'C'], [1, 2, 3])
df = pd.get_dummies(df)
```

```python
df
```

```python
# afficher un résumé des données manquantes

print(df.isnull().sum())
```

```python
# On remplace les NAN par des 1, 2, 3 de manière aléatoire
from random import randint

c = randint(1,3)
# On remplace les NAN 
df = df.fillna(c)
print(df.isnull().sum())
print(c)
```
